```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('/content/sample_data/cancerpatientdatasets.xlsx')
```

```python
df.head()
```

```python
def tran_cat_to_num(df):
    if df['Gender'] == 1:
        return 'male'
    else:
        return 'female'
```

```python
# create sex 
df['sex1'] = df.apply(tran_cat_to_num, axis=1)
df
```

```python
df.head()
```

```python
men_women = df.groupby('sex1').sum() / df.groupby('sex1').sum().sum()
men_women
```

```python
q_men_women = df['sex1'].value_counts()
q_men_women
```

```python
plt.pie(q_men_women, labels=['Men', 'Women'], autopct='%1.1f%%', shadow=True);
```

```python
q_level = df['Level'].value_counts()
q_level
```

```python
plt.pie(q_level, labels=['High', 'Medium', 'Low'], autopct='%1.1f%%', shadow=True);
```
